# 北尾早霧・砂川武貴・山田知明『定量的マクロ経済学と数値計算』日本評論社
## 第2章：2期間モデルと数値計算の概観

```julia
rent = 1.025^30 - 1.0
```

```julia
# 予算制約を描いてみよう
w = 0.5
nw = 101
c1 = collect(LinRange(0, w, nw));
```

```julia
c2 = (1+rent).*(w .- c1);
```

```julia
using Plots
pyplot()
```

```julia
plt = plot(c1, c2,
    color = :blue,
    legend = :none,
    xlims = (0, 1.0),
    ylims = (0, 1.5),
    xlabel = ("若年期の消費"),
    ylabel = ("老年期の消費"),
    linewidth = 4,
    markersize = 8,
    titlefont = font("HackGen35Nerd", 12),
    guidefont = font("HackGen35Nerd", 12),
    tickfont = font("HackGen35Nerd", 8),
    framestyle = :semi
)
```
